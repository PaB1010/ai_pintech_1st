```python
from faker import Faker
```

```python
# 기본 값은 영어 데이터
faker = Faker()

# 호출시마다 랜덤하게 데이터 변경
faker.name()
```

```python
# 한국어 데이터 생성
faker = Faker('ko-Kr')

faker.name()
```

```python
# Tuple형태로 1000번 반복한 List 데이터 생성
members = [(faker.email(), faker.name(), faker.address())
          for i in range(0, 1000)]
# i 변수가 사용되지 않는 경우 자리만 채워주기 (_)
```

```python
for member in members : 
    print(member)
```
